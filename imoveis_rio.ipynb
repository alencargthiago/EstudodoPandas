## Revisando conceitos da Biblioteca Pandas utilizando dados do Spotify e youtube 
Entendendo as funcionalidades do pandas. Primeiros passos utilizando o dataset __inserir aqui__

Alguns conceitos são fundamentais para o melhor entedimento das funcionalidades da biblioteca pandas:

**DataFrame**: Por definição, um DataFrame é uma estrutura de dados bidimensional tabular que possui  rótulos nas linhas e colunas. Semelhante a uma tabela SQL ou uma planilha do Excel. DataFrames podem ser criados a partir de várias fontes de dados, como arquivos CSV, Excel, JSON, SQL e outros.

**Series**: Uma Series é semelhante a uma coluna de uma planilha do Excel ou uma tabela SQL. 

**Indexação e seleção**: O Pandas oferece várias maneiras de selecionar e indexar dados em um DataFrame ou uma Series, como iloc, loc, e indexação booleana. Isso permite que você selecione linhas, colunas ou células específicas em um DataFrame ou valores específicos em uma Series.

**Leitura e escrita de arquivos**: O Pandas fornece funções para ler e escrever dados em diferentes formatos, como CSV, Excel, JSON, HTML e SQL, tais como: read_csv, read_excel, to_csv e to_excel etc.

**Manipulação de dados**: Pandas fornece uma ampla variedade de funções para limpar, transformar e manipular dados. Por exemplo, você pode usar funções como drop, rename, sort_values, groupby, merge, concat, pivot e outras para modificar os dados conforme necessário.

**Análise de dados** Também é possível realizar análises estatísticas em seus dados. Algumas funções incluem __describe, mean, median, min, max, std, corr__ e muitas outras.

**Tratamento de dados ausentes**: Pandas fornece funções para lidar com dados ausentes ou nulos, como isnull, __notnull, dropna e fillna__.

Ainda pode-se utilizar o método pd.pivolt_table() para agrupar nossos dados. É importante ter em mente que:
- pd.pivot_table(): permite criar uma tabela de resumo com base em várias colunas do DataFrame original, agregando os dados usando uma função (como média, soma, contagem, etc.) e apresentando os resultados em um formato tabular hierárquico.

- head(): é um método simples que retorna as primeiras 'n' linhas do DataFrame. Ele não faz nenhum tipo de transformação ou agregação de dados.

- Métodos de filtragem: .filter, .mask, e .query são usados para filtrar os dados de um DataFrame de acordo com determinadas condições. Eles não agregam os dados, apenas retornam um subconjunto do DataFrame original com base nos critérios de filtragem.

- groupby(): é usado para agrupar os dados com base em uma ou mais colunas. Ele cria um objeto "GroupBy", que pode ser usado em conjunto com funções de agregação (como .mean(), .sum(), etc.) para calcular estatísticas resumidas para cada grupo.

- A escolha entre essas funções depende do objetivo da análise. Se você deseja apenas filtrar os dados com base em critérios específicos, os métodos de filtragem são mais apropriados. Se você deseja agregar os dados e calcular estatísticas de resumo, .groupby() e pd.pivot_table() são mais úteis. A principal diferença entre groupby() e pivot_table() é a forma como os dados são apresentados: groupby() retorna um DataFrame "achatado", enquanto pivot_table() retorna uma tabela hierárquica.

```python
import numpy as np
pd.pivot_table(spotify_youtube, index="Artist", aggfunc=np.mean)
```

```python
pd.pivot_table(spotify_youtube, index=['Artist'], values =['Stream','Likes','Danceability','Energy'], aggfunc=np.mean)
```

os dados são agrupados por 'Artist'. As colunas 'Stream', 'Likes', 'Danceability' e 'Energy' são agregados de acordo com a média

## Missing e Alterações/Criações de Campos

O método df.isna() verifica verdadeiro ou falso para missing.

```python
spotify_youtube.isna() #verifica se true dados missing
```

```python
spotify_youtube.isna().sum() #contabiliza quantos missing por coluna
```

```python
spotify_youtube.info() #informação sobbre dataset e dados nulos, ele conta caso exista.
```

```python
spotify_youtube[pd.isna(spotify_youtube['Artist'])] #indicaria linhas com campo Artista missing
```

```python
spotify_youtube[spotify_youtube['Artist'].notna()] #filtra as linhas com as colunas Artista não ausentes
```

Também pode-se substituir os missings por outros valores utilizando


```python
spotify_youtube.fillna('0') #substitue os missings por zero.
```

Também podemos substituir missings por medidas estatísticas, como a média através do método __df.fillna(df.mean(), inplace= true)__

É improtante ressaltar que o método inplace= true irá substituir os missing no dataframe original e não na copia, por isso é importante ter muito CUIDADO!

Também podemos criar um dicionário para substituir colunas do data frames da seguinte maneira


```python
values= {'Artist':'desconhecido','Stream':'Flopado'} #caso exista artista missings, sera substituindo por desconhecido
spotify_youtube.fillna(value=values)                 #caso exista stream missing, sera substituidom por flopado
```

O método __df.dropna()__ irá excluir todas as linhas que possui valores __nulos (missings)__, contudo, se alguma linha do seu DataFrame contiver um valor nulo em qualquer uma das colunas, essa linha será excluída ao usar **df.dropna()** Por isso é importante ter muita cautela

```python
spotify_youtube.dropna() #exclui todas as linhas com valores nulos
```

```python
spotify_youtube.dropna(subset=['Artist']) #exclui apenas linhas com missings na coluna Artist
```

## Alterando e criando campos


As vezes pode ser necessario trocar valores dentro de campos especificios, uma boa forma de fazer isso é através de funções. Neste caso, criou-se nova coluna onde substituimos os valores de True para Sim. É melhor fazer com valores númericos, então para atingir o objetivo, foi preciso converter os valores da coluna licensed para **str** através do método **astype(str)**

```python
def substitue_true(value):
    x= value.replace('True', 'Sim')
    return str (x)
spotify_youtube['Licensed_new']=spotify_youtube['Licensed'].astype(str).apply(substitue_true)
spotify_youtube.head()
```

```python
spotify_youtube['Licensed_new'].info() #informções sobre a nova coluna
```

```python
spotify_youtube['Dance_energy']=(spotify_youtube['Danceability']+spotify_youtube['Energy'])/2
spotify_youtube.head()
```
